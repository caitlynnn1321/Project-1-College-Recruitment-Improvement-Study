<h5 style="font-weight: bold; font-family: Arial; font-size: 15px;">
    Question #1: Which platforms/events are the best for advertising computer science courses?
</h5>


<h5 style=" font-family: Arial; font-size: 15px;">
 Survey Questions: 3, 5, & 8
</h5>


```python
# Define the columns for each question
q3_cols = ['precollege_open_house', 'precollege_instant_decision', 
           'precollege_info_session', 'precollege_titans_tuesday', 
           'precollege_women_who_dare', 'precollege_college_fair']

q5_cols = ['heard_website', 'heard_billboard', 'heard_tv', 'heard_radio', 
           'heard_social_media', 'heard_event', 'heard_family_friend', 
           'heard_employer', 'heard_current_student', 'heard_alumni']

q8_cols = ['info_high_school_counselor', 'info_it_website', 'info_admissions', 
           'info_ccm_advisor', 'info_employer', 'info_other']

# Calculate aggregated effectiveness for each platform
# (combining all three questions into one overall score)
platforms_effectiveness = {
    'On-Campus\nSession': [
        (df['precollege_info_session'] == 'Yes').sum(),
        (df['heard_event'] == 'Yes').sum()
    ],
    'Website': [
        (df['heard_website'] == 'Yes').sum(),
        (df['info_it_website'] == 'Yes').sum()
    ],
    'College Fair': [
        (df['precollege_college_fair'] == 'Yes').sum(),
        (df['heard_event'] == 'Yes').sum()
    ],
    'Employer': [
        (df['heard_employer'] == 'Yes').sum(),
        (df['info_employer'] == 'Yes').sum()
    ],
    'Family/Friend': [
        (df['heard_family_friend'] == 'Yes').sum()
    ],
    'Current Student': [
        (df['heard_current_student'] == 'Yes').sum()
    ],
    'HS Counselor': [
        (df['info_high_school_counselor'] == 'Yes').sum()
    ],
    'Admissions': [
        (df['info_admissions'] == 'Yes').sum()
    ],
    'Social Media': [
        (df['heard_social_media'] == 'Yes').sum()
    ],
    'Alumni': [
        (df['heard_alumni'] == 'Yes').sum()
    ]
}

# Calculate total "Yes" responses and average effectiveness for each platform
total_responses = len(df)
platform_names = []
effectiveness_scores = []

for platform, responses in platforms_effectiveness.items():
    total_yes = sum(responses)
    avg_score = (total_yes / total_responses) * 100
    platform_names.append(platform)
    effectiveness_scores.append(avg_score)

# Sort by effectiveness
sorted_data = sorted(zip(platform_names, effectiveness_scores), key=lambda x: x[1], reverse=True)
platform_names, effectiveness_scores = zip(*sorted_data)

# Create bar chart
fig, ax = plt.subplots(figsize=(12, 7))

colors = ['#2E86AB' if i < 3 else '#A23B72' if i < 6 else '#F18F01' 
          for i in range(len(platform_names))]
bars = ax.barh(platform_names, effectiveness_scores, color=colors)

ax.set_xlabel('Effectiveness Score (% of Respondents)', fontsize=12, fontweight='bold')
ax.set_ylabel('Platforms/Channels', fontsize=12, fontweight='bold')
ax.set_title('Best Platforms/Events for Advertising Computer Science Courses\n(Aggregated from Survey Questions 3, 5, and 8)', 
             fontsize=14, fontweight='bold', pad=20)

# Add value labels on bars
for bar in bars:
    width = bar.get_width()
    ax.text(width, bar.get_y() + bar.get_height()/2.,
           f'{width:.1f}%',
           ha='left', va='center', fontsize=10, fontweight='bold')

ax.set_xlim(0, max(effectiveness_scores) * 1.15)
ax.grid(axis='x', alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()
```

<h5 style="font-family: Arial; font-size: 15px;">
Analysis: Word-of-mouth remains the most effective recruitment channel at 76.1%, with family and friend referrals significantly outpacing other platforms. The CCM website and current student recommendations follow closely at 60.9% and 50.0% respectively, indicating that both digital presence and peer advocacy are critical for reaching prospective CS students. On-campus information sessions (38.0%), social media (28.3%), and alumni networks (28.3%) provide strong secondary touchpoints. Notably, high school counselors represent an underutilized opportunity at only 10.9%, suggesting significant potential for growth through enhanced K-12 partnerships. Overall, a integrated approach leveraging personal networks, digital channels, and direct engagement—paired with targeted outreach to high school guidance departments—would maximize the program's advertising effectiveness.</h5>


<h5 style="font-weight: bold; font-family: Arial; font-size: 15px;">
    Question #2: What are the most common motivations for students to enroll in a computer science course?
</h5>


<h5 style=" font-family: Arial; font-size: 15px;">
 Survey Questions: 2
</h5>


```python
# Define the motivation columns from Question 2
motivation_cols = {
    'Required for Degree': 'motivation_required_degree',
    'Keep Skills Current': 'motivation_keep_current',
    'Career Advancement': 'motivation_career_advancement',
    'Career Change': 'motivation_career_change',
    'Professional Development': 'motivation_professional_dev',
    'Job Displacement': 'motivation_job_displacement',
    'Relocation': 'motivation_relocation',
    'IT Industry Certifications': 'motivation_it_certifications',
    'Financial': 'motivation_financial',
    'Personal Enrichment': 'motivation_personal_enrichment',
    'Curiosity': 'motivation_curiosity'
}

# Calculate percentage of "Yes" responses for each motivation
total_responses = len(df)
motivations = []
percentages = []

for motivation_name, col_name in motivation_cols.items():
    if col_name in df.columns:
        yes_count = (df[col_name] == 'Yes').sum()
        percentage = (yes_count / total_responses) * 100
        motivations.append(motivation_name)
        percentages.append(percentage)

# Sort by percentage in descending order
sorted_data = sorted(zip(motivations, percentages), key=lambda x: x[1], reverse=True)
motivations, percentages = zip(*sorted_data)

# Create horizontal bar chart
fig, ax = plt.subplots(figsize=(12, 8))

colors = plt.cm.viridis(np.linspace(0.3, 0.9, len(motivations)))
bars = ax.barh(motivations, percentages, color=colors)

ax.set_xlabel('Percentage of Respondents (%)', fontsize=12, fontweight='bold')
ax.set_ylabel('Motivation', fontsize=12, fontweight='bold')
ax.set_title('Student Motivations for Enrolling in Computer Science Courses\n(% of Students Who Selected Each Reason)', 
             fontsize=14, fontweight='bold', pad=20)

# Add value labels on bars
for bar in bars:
    width = bar.get_width()
    ax.text(width, bar.get_y() + bar.get_height()/2.,
           f'{width:.1f}%',
           ha='left', va='center', fontsize=10, fontweight='bold')

ax.set_xlim(0, max(percentages) * 1.15)
ax.grid(axis='x', alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()
```

<h5 style="font-family: Arial; font-size: 15px;">
Analysis: The most common motivation for students to enroll in computer science courses is keeping skills current at 76.1%, indicating that professional development and staying competitive in the tech industry are primary drivers. Career advancement follows closely at 58.7%, demonstrating that students view CS education as a pathway to professional growth. Curiosity ranks third at 51.1%, suggesting a significant portion of students are intrinsically motivated to learn computing concepts. Practical concerns also play a role, with job displacement (22.8%), financial considerations (21.7%), and career change (19.6%) motivating approximately one-fifth of students. Notably, relocation (9.8%) is the least common motivation. These findings suggest that the program attracts both career-focused professionals seeking to maintain competitiveness and exploratory learners interested in expanding their knowledge, with a smaller but meaningful cohort driven by life circumstances such as job loss or geographic relocation.</h5>


<h5 style="font-weight: bold; font-family: Arial; font-size: 15px;">
    Question #3: Which demographic is interested in taking more Computer Science, Information Technologies, and Game Development than others? 
    
</h5>



<h5 style=" font-family: Arial; font-size: 15px;">
 Survey Questions: 9, 11, 12, 13
</h5>

```python

# Convert interest_level to numeric (1-5 scale)
df['interest_level'] = pd.to_numeric(df['interest_level'], errors='coerce')

# Calculate average interest level by Gender
gender_interest = df.groupby('gender')['interest_level'].agg(['mean', 'count']).reset_index()
gender_interest.columns = ['Demographic', 'Avg Interest', 'Count']
gender_interest['Category'] = 'Gender'

# Calculate average interest level by Age
age_interest = df.groupby('age')['interest_level'].agg(['mean', 'count']).reset_index()
age_interest.columns = ['Demographic', 'Avg Interest', 'Count']
age_interest['Category'] = 'Age'

# Calculate average interest level by Race/Ethnicity
race_interest = df.groupby('race_ethnicity')['interest_level'].agg(['mean', 'count']).reset_index()
race_interest.columns = ['Demographic', 'Avg Interest', 'Count']
race_interest['Category'] = 'Race/Ethnicity'

# Combine all demographic data
all_demographics = pd.concat([gender_interest, age_interest, race_interest], ignore_index=True)

# Remove rows with NaN or no data
all_demographics = all_demographics.dropna(subset=['Avg Interest'])
all_demographics = all_demographics[all_demographics['Count'] > 0]

# Create single comprehensive figure
fig, axes = plt.subplots(3, 1, figsize=(14, 14))
fig.suptitle('CS/IT/Game Development Interest by Demographic Groups\n(Average Interest Rating on 1-5 Scale, n = number of respondents)', 
             fontsize=16, fontweight='bold', y=0.995)

categories = ['Gender', 'Age', 'Race/Ethnicity']
colors_list = ['#2A9D8F', '#264653', '#E76F51']

for idx, (ax, category, color) in enumerate(zip(axes, categories, colors_list)):
    data = all_demographics[all_demographics['Category'] == category].sort_values('Avg Interest', ascending=False)
    
    # Only show groups with n >= 2 for better statistical validity
    data_filtered = data[data['Count'] >= 2]
    
    bars = ax.bar(range(len(data_filtered)), data_filtered['Avg Interest'], color=color, alpha=0.8, edgecolor='black', linewidth=1.5)
    
    ax.set_ylabel('Average Interest Level', fontsize=11, fontweight='bold')
    ax.set_title(f'{category}', fontsize=13, fontweight='bold', loc='left')
    ax.set_ylim(0, 5)
    ax.set_xticks(range(len(data_filtered)))
    ax.set_xticklabels(data_filtered['Demographic'], rotation=45, ha='right', fontsize=10)
    ax.axhline(y=3, color='gray', linestyle='--', alpha=0.5, linewidth=2, label='Neutral (3.0)')
    ax.grid(axis='y', alpha=0.3, linestyle='--')
    ax.set_axisbelow(True)
    
    # Add value labels on top of bars with sample size
    for bar, count, interest in zip(bars, data_filtered['Count'], data_filtered['Avg Interest']):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + 0.15,
               f'{interest:.2f}\n(n={int(count)})',
               ha='center', va='bottom', fontsize=10, fontweight='bold')
    
    if idx == 0:
        ax.legend(loc='upper right', fontsize=10)

plt.tight_layout()
plt.show()
```

<h5 style="font-family: Arial; font-size: 15px;">
Analysis: Gender analysis reveals minimal differences in CS interest, with women (2.62) and men (2.59) showing nearly identical engagement levels around the neutral midpoint (3.0), though the sample sizes are relatively balanced. Age demographics demonstrate a more pronounced pattern, with students aged 25-34 showing the highest interest (3.18), followed by those 18 and younger (2.86) and 19-20 year-olds (2.30), suggesting that career-established adults are most motivated to pursue additional CS coursework. Race/ethnicity results highlight significant variation, with students identifying as Hispanic or Latino/Asian/White/Caucasian (combined) showing exceptional interest (5.00), though this group has very limited representation (n=1). Among groups with more substantial sample sizes, those identifying as Asian (3.33, n=3) and multi-ethnic respondents (3.40, n=5) demonstrate above-neutral interest, while White/Caucasian students (2.88, n=36) represent the largest demographic but show below-neutral engagement. These findings suggest that mature learners (25-34) and certain underrepresented minorities may be particularly strong targets for CS program recruitment, while traditional college-age students (19-20) and the largest demographic cohort show relatively lower interest levels.</h5>


<h5 style="font-weight: bold; font-family: Arial; font-size: 15px;">
    Question #4: Which degree programs have the most students enrolled in computer science classes, and what are the most commonly taken computer science courses within those programs?
</h5>

<h5 style=" font-family: Arial; font-size: 15px;">
 Survey Questions: 1 & 7
</h5>

```python
# Count students by degree program
program_counts = df['degree_program'].value_counts().sort_values(ascending=False)

# Get top 8 degree programs
top_programs = program_counts.head(8).index.tolist()

# Filter data for top programs only
df_top = df[df['degree_program'].isin(top_programs)]

# Create a cross-tabulation of degree programs and courses
course_by_program = pd.crosstab(df_top['degree_program'], df_top['current_course'])

# Sort by total students per program (using only top 8)
course_by_program = course_by_program.loc[program_counts[top_programs].index]

# Create stacked bar chart
fig, ax = plt.subplots(figsize=(18, 8))

# Plot with new colors: pink, burgundy, orange
course_by_program.plot(kind='barh', stacked=True, ax=ax, 
                        color=['#FF6B9D', '#C44569', '#FFA502'],
                        edgecolor='black', linewidth=2.5)

ax.set_xlabel('Number of Students', fontsize=12, fontweight='bold')
ax.set_ylabel('Degree Program', fontsize=12, fontweight='bold')
ax.set_title('Computer Science Course Enrollment by Degree Program\n(Showing Top 8 Programs with Highest CS Enrollment)', 
             fontsize=14, fontweight='bold', pad=20)

ax.legend(title='Course', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10, title_fontsize=11)
ax.grid(axis='x', alpha=0.3, linestyle='--')
ax.set_axisbelow(True)

# Add total count labels well to the right of each bar
for i, program in enumerate(course_by_program.index):
    total = course_by_program.loc[program].sum()
    ax.text(total + 1.5, i, f'n={int(total)}', va='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

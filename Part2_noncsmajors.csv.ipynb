<h5 font-weight:bold; font-family:Arial;">Cleaning Categorical Values</h5>


<h5 font-family:Arial;">Organizing the degree programs and creating new categories where necessary.</h5>

```python
df['degree_program'].unique()
```

```python
major_mapping = {
    # --- Direct matches from the form ---
    'Broadcasting Arts and Technology': 'Broadcasting Arts and Technology',
    'Child and Family Studies': 'Child and Family Studies',
    'Communication': 'Communication',
    'Criminal Justice': 'Criminal Justice',
    'Early Childhood Education': 'Early Childhood Education',
    'Exercise Science': 'Exercise Science',
    'Journalism': 'Journalism',
    'Liberal Arts': 'Liberal Arts',
    'Music Technology': 'Music Technology',
    'Nursing': 'Nursing',
    'Radiography': 'Radiography',
    'Not in a degree program': 'Not in a degree program',

    # --- Slight variations or clarifications ---
    'Nursing, but not at CCM - I am a visiting student ': 'Nursing',
    'Musical Theatre': 'Music Technology',  # fits under same creative arts area

    # --- Related / similar but not exact form options ---
    'Engineering Science': 'Engineering',
    'Biology': 'Math / Science',
    'Biology - Environmental Track': 'Math / Science',
    'Culinary Arts and Sciences': 'Hospitality / Culinary',
    'History Education': 'Liberal Arts',
    'Psychology': 'Psychology / Social Science',
    'child psychology ': 'Psychology / Social Science',
    'analyst': 'Computer / Technology',  # could indicate Data / IT-related field

    # --- Default fallback ---
    'Other': 'Other'
}
```

```python
df['major_mapped'] = df['degree_program'].map(major_mapping)
```
